## Задание 3

Производится 8 выстрелов по цели, в каждом из которых вероятность попадания равна 0.3. Найдите вероятность того, что цель будет поражена хотя бы два раза.

Ответ округлите до 3 знаков после запятой.

```python
p3 = 0 
for n in range(2,8+1):
    p3 += special.comb(8,n)*(0.3)**n*(1-0.3)**(8-n)
    
print(p3.round(3))
```

## Задание 4

## Задание 5

В телеателье имеется 7 телевизоров. Для каждого телевизора вероятность того, что в данный момент он включен, равна 0.5. Найдите вероятность того, что в данный момент включены не менее трех телевизоров.

Ответ округлите до 3 знака.

```python
p5 = 0 
p_pol = 0.5
m = 7
k = 3
for n in range(k,m+1):
    p5 += special.comb(m,n)*(p_pol)**n*(1-p_pol)**(m-n)
    
print(round(p5, 3))
```

## Задание 7

```python
p7 = 0 
p_pol = 1 - 0.1
m = 10
k = 8
for n in range(k,m+1):
    p7 += special.comb(m,n)*(p_pol)**n*(1-p_pol)**(m-n)
    
print(round(p7, 2))
```
