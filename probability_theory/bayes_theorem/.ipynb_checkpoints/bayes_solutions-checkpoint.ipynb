## Задание 1

Из 500 компьютеров  180 принадлежат к 1 партии, 170 – ко второй партии, остальные к третьей. В первой партии 3% брака, во второй — 2%, в третьей – 6%. Случайно выбирается один компьютер. Определите вероятность того, что выбранный компьютер – бракованный. 

Округлите полученный ответ до 4 знаков. В качестве разделителя здесь и далее используйте точку.



```python
n = 500
n1, p1 = 180, 0.03
n2, p2 = 170, 0.02
n3, p3 = n - n1 - n2, 0.06
p_brak = n1/n * p1 + n2/n * p2 + n3/n * p3
print(round(p_brak,4))
```

## Задание 2

Есть компания с программистами. В ней на языке Java пишут в 3 раза больше человек, чем на языке C++. Программистов на языке C++ в 4 раза больше, чем на Python.

Вероятность сделать работу за один день для Java-программиста — 0.85, для программиста на языке C++ 0.9, а для Python программиста 0.8. Найти вероятность того, что программист, выбранный наугад, сделает работу за один день.

Ответ округлите до 2 знака. 

```python
n_j, p_j = 12/17 , 0.85
n_c, p_c = 4/17 , 0.9
n_p, p_p = 1 - n_j - n_c, 0.8
p_1_day = n_j * p_j + n_c * p_c + n_p * p_p
print(round(p_1_day,2))
```

## Задание 3

Из 40 снайперов:

- 18 попадает в цель с вероятностью 0.9
- 8 — с вероятностью 0.4 
- 14 – с вероятностью 0,7
Случайно выбранный снайпер произвел выстрел, поразив цель. Вычислите, снайпер из какой группы наиболее вероятно совершил этот выстрел. 

В ответ запишите наибольшую вероятность, округленную до 3 знака.

```python
n_all = 40
n = [18, 8, 14]
p = [0.9, 0.4, 0.7]
p_all = n[0]/n_all * p[0] + n[1]/n_all * p[1] + n[2]/n_all * p[2]
for i in range(3):
    p_i = (n[i]/n_all)*p[i]/ p_all
    print(i+1, ' группа: ', round(p_i,3))
```

## Задание 5

На гитхабе лежит 12 скриптов, написанных аналитиком №1, 20 скриптов — написанных аналитиком №2 и 18 скриптов — написанных аналитиком №3.

Вероятность того, что скрипт, написанный аналитиком №1, работает без ошибок, равна 0.9; для скриптов, написанных аналитиками №2 и №3, эти вероятности соответственно равны 0.6 и 0.9.

Нужно найти вероятность того, что случайно взятый скрипт будет работать без ошибок. В ответ запишите число, округленное до 2 знака.

```python
n = [12, 20, 18]
n_all = n[0] + n[1] + n[2]
p = [0.9, 0.6, 0.9 ]
p_all = n[0]/n_all * p[0] + n[1]/n_all * p[1] + n[2]/n_all * p[2]
round(p_all, 2)
```

## Задание 6

Известно, что около 60% всех людей в городе болеют заболеванием А. Для определения того, заболел человек или нет, проводится тест, точность которого составляет 99%. Вы получили положительный тест, какова вероятность того, что вы реально больны?

Округлите ответ до 3 знака.



```python
n = [0.6, 0.4]
p = [0.99, 0.01]
p_all = n[0]/n_all * p[0] + n[1]/n_all * p[1]
p_ill = n[0]/n_all * p[0] / p_all
round(p_ill, 3)
```

В любой момент времени вероятность того, что самолет находится в зоне действия радара, составляет 7%.

Какова вероятность того, что самолет находится в зоне действия радара, учитывая, что на радаре нет самолета?

Ответ округлите до 4 знаков.

```python
n = [0.07, 0.93]
p = [0.95, 0.05]
p_all = n[0] * p[0] + n[1] * p[1] 
round(n[1] * p[1] / p_all, 4)
```
